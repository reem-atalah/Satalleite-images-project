```python
import cv2
import os

# loop for flooded and non-flooded classes
classes = ['flooded', 'non-flooded']

for className in classes:
    # Set the directory paths
    input_dir = "/content/drive/MyDrive/SI_flood_dataset/dataset/SI_flood_dataset_split/train_resized_eq/" + className + "/"
    output_dir = "/content/drive/MyDrive/SI_flood_dataset/dataset/SI_flood_dataset_split/train_aug_resized_eq/" + className + "/"

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Loop through all the files in the input directory
    for filename in os.listdir(input_dir):
        # Read the image
        img = cv2.imread(os.path.join(input_dir, filename))
        # Rotate the image by 90 degrees clockwise
        rotated_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
        rotated_180 = cv2.rotate(img, cv2.ROTATE_180)
        # Save the rotated image to the output directory
        # split the name from the extension
        filename = os.path.splitext(filename)[0]
        cv2.imwrite(os.path.join(output_dir, filename+'_normal.jpg'), img)
        cv2.imwrite(os.path.join(output_dir, filename+'_90.jpg'), rotated_90)
        cv2.imwrite(os.path.join(output_dir, filename+'_180.jpg'), rotated_180)
        print(filename)
```
